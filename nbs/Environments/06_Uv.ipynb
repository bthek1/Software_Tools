Designed to replace tools like `pip`, `virtualenv`, `pyenv`, and `poetry`, `uv` offers a unified interface for managing Python environments, dependencies, and versions. **Example:**

**Example:**
```bash
uv remove flask
```

### 4. **`uv sync`**îˆƒSynchronizes the project's environment with the dependencies specified in `pyproject.toml`

**Example:**
```bash
uv sync
```

### 5. **`uv lock`**îˆƒGenerates or updates the `uv.lock` file, locking the exact versions of dependencies

**Example:**
```bash
uv lock
```

### 6. **`uv run <script.py>`**îˆƒRuns a Python script within the project's environment

**Example:**
```bash
uv run app.py
```

### 7. **`uv python install <version>`**îˆƒInstalls a specific Python version for the project

**Example:**
```bash
uv python install 3.11
```

### 8. **`uv python use <version>`**îˆƒSets the project's Python version to the specified version

**Example:**
```bash
uv python use 3.11
```

### 9. **`uv pip install <package>`**îˆƒInstalls a package using `pip` within the project's environment

**Example:**
```bash
uv pip install requests
```

### 10. **`uv pip uninstall <package>`**îˆƒUninstalls a package using `pip` within the project's environment

**Example:**
```bash
uv pip uninstall requests
```
