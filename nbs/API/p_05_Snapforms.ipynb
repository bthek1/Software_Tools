# Snap Forms Integration

## Authenticate to the API

```python
import os
from dotenv import load_dotenv
import requests
import pandas as pd 
# Load environment variables from .env file
load_dotenv()

# Get environment variables
client_id = os.getenv('Snapforms_Client_ID')
client_secret = os.getenv('Snapforms_Client_secret')
username = os.getenv('Snapforms_username')
password = os.getenv('Snapforms_password')
```

```python

url = "https://user.snapforms.com.au/oauth/token"

# Prepare the payload using the environment variables
payload = {
    'client_id': client_id,
    'client_secret': client_secret,
    'grant_type': 'password',
    'username': username,
    'password': password
}

headers = {
    'Content-Type': 'application/x-www-form-urlencoded'
}

# Make the request
auth_response = requests.post(url, headers=headers, data=payload)

# Check for successful response
if auth_response.status_code == 200:
    print("Access Token:", auth_response.json().get("access_token"))
else:
    print("Failed to retrieve access token:", auth_response.text)
```

```python
auth_json = auth_response.json()
auth_json
```

## Get Form Names

```python
import requests

url = "https://user.snapforms.com.au/api/forms/"
token = auth_json["access_token"]  # Replace with your actual token

headers = {
    "Authorization": f"Bearer {token}"
}

# Send the GET request with Authorization header
forms_response = requests.get(url, headers=headers)

# Check for successful response
if forms_response.status_code == 200:
    print("Success:", forms_response.json())  # If response is JSON, print as JSON
else:
    print("Failed:", forms_response.status_code, forms_response.text)
```

```python
forms_json = forms_response.json()
forms_pd = pd.DataFrame(forms_json)
forms_pd
```

```python
questionnaire_1 = forms_json[0]['name']
```

```python
# Create an empty list to store rows
data = []

# Iterate over each response and its answers
for response in response_json.get('responses', []):
    response_id = response.get('response_id')
    datetime = response.get('datetime')
    datetime_modified = response.get('datetime_modified')
    
    for answer in response.get('answers', []):
        question = answer.get('question')
        answer_type = answer.get('type')
        answer_value = answer.get('answer')
        
        # Append the row as a dictionary
        data.append({
            "response_id": response_id,
            "datetime": datetime,
            "datetime_modified": datetime_modified,
            "question": question,
            "answer_type": answer_type,
            "answer_value": answer_value
        })

# Create a DataFrame from the list of dictionaries
df = pd.DataFrame(data)
df
```
